```python
import matplotlib.pyplot as plt

cache_sizes_kb = ['32 KB', '64 KB', '128 KB', '256 KB', '512 KB', '1 MB', '2 MB', '4 MB', '8 MB', '16 MB', '32 MB', '64 MB', '128 MB']

areas_with_associativity_4 = [0.902279, 1.27551, 2.02069, 3.49898, 8.33802, 18.0265, 31.8043, 64.983, 116.672, 226.253, 428.905, 861.726, 1681.97]

associativities = [1, 2, 4, 8, 16, 32, 64]
associativities = [str(val) for val in associativities]
areas_with_cache_size_128KB = [1.77945, 1.78124, 2.02069, 2.53908, 7.30267, 15.3807, 32.8221]

plt.figure(figsize=(10,6))
plt.plot(cache_sizes_kb, areas_with_associativity_4, marker='o', label='Area vs Cache Size')
plt.xlabel('Cache Size')
plt.ylabel('Area (mm^2)')
plt.title('Impact of Cache Size on Area of Data Array with Associativity 4')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.plot(associativities, areas_with_cache_size_128KB, marker='o', label='Area vs Associativity')
plt.xticks(associativities)
plt.xlabel('Associativity')
plt.ylabel('Area (mm^2)')
plt.title('Impact of Associativity on Area of Data Array with Cache Size 128 KB')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
